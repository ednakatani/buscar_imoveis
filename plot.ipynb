```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('imoveis_zap.csv', sep=',')
df.head()
```

```python
correlacoes = df.corr()
fig, ax = plt.subplots()
ax.set_title('Correlação entre as variáveis')
sns.heatmap(correlacoes,square=True, annot=True)
plt.show()
```
