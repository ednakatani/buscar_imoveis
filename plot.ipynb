```python
import pandas as pd
```

```python
df = pd.read_csv('distribuicao_imoveis.csv', sep=';')
df.head()
```

```python
plot_df = pd.DataFrame({
   'ocorrencia': df.quantidade
   }, index=df.bairro)

ax = plot_df.plot(figsize=(20,3), title='Ocorrência de Imóveis por Bairro em Curitiba')

ax.set_xticks(df.bairro)
ax.axhline(y=df.quantidade.mean(), color='r', linestyle='--')
ax.set_facecolor('w')

fig = ax.get_figure()
fig.savefig('plot.png')
```
